```python
matches['matchup'] = matches.apply(
    lambda row: ' vs '.join(sorted([row['team1'], row['team2']])), axis=1
)

match_counts = matches['matchup'].value_counts()

winner_counts = matches.groupby(['matchup', 'winner']).size().reset_index(name='win_count')

most_wins = winner_counts.sort_values(['matchup', 'win_count'], ascending=[True, False]) \
                         .drop_duplicates(subset='matchup', keep='first')

final_df = most_wins.merge(match_counts.rename('total_matches'), left_on='matchup', right_index=True)

final_df = final_df.sort_values('total_matches', ascending=False).reset_index(drop=True)

print(final_df.head(5))  
```

```python
# What are the top-scoring batsmen by total runs, strike rate, and boundary count

valid_balls = delivery[~delivery['extra_runs'].gt(0)]

# Aggregate stats
batsman_stats = delivery.groupby('batter').agg(
    total_runs=('batsman_runs', 'sum'),
    balls_faced=('ball', 'count'),
    fours=('batsman_runs', lambda x: (x == 4).sum()),
    sixes=('batsman_runs', lambda x: (x == 6).sum())
).reset_index()

# Calculate strike rate and boundaries
batsman_stats['strike_rate'] = (batsman_stats['total_runs'] / batsman_stats['balls_faced']) * 100
batsman_stats['boundaries'] = batsman_stats['fours'] + batsman_stats['sixes']

# Filter out batsmen who played very few balls (e.g., < 30 balls)
batsman_stats = batsman_stats[batsman_stats['balls_faced'] > 30]

# Sort by multiple metrics
top_batsmen = batsman_stats.sort_values(
    by=['total_runs', 'strike_rate', 'boundaries'],
    ascending=[False, False, False]
)

# Display top 10
print(top_batsmen.head(10))
```

```python
# Who are the best death-over bowlers (Overs 17-20)
death_overs=delivery[delivery['over'].between(17,20)]
death_bowlers=death_overs.groupby('bowler').agg(
    balls=('ball','count'),
    runs=('total_runs','sum'),
    wickets=('is_wicket','sum')
).reset_index()

death_bowlers
```

```python
# Which team has the best economy rate across all seasons?
death_overs = delivery[delivery['over'].between(17, 20)]

death_bowlers = death_overs.groupby('bowler').agg(
    balls=('ball', 'count'),
    runs=('total_runs', 'sum'),
    wickets=('is_wicket', 'sum')
).reset_index()

death_bowlers['economy'] = death_bowlers['runs'] / (death_bowlers['balls'] / 6)

best_death_bowlers = death_bowlers.sort_values(['wickets', 'economy'], ascending=[False, True])
best_death_bowlers
```

```python
# Combine matches and deliveries data to calculate match-wise total team runs and compare with target scores

team_runs = delivery.groupby(['match_id', 'inning', 'batting_team'])['total_runs'].sum().reset_index()
team_runs.rename(columns={'total_runs': 'team_total'}, inplace=True)

# Step 2: Split 1st and 2nd innings
first_innings = team_runs[team_runs['inning'] == 1]
second_innings = team_runs[team_runs['inning'] == 2]

# Step 3: Merge first and second innings on match_id
combined = pd.merge(first_innings, second_innings, on='match_id', suffixes=('_target', '_chased'))

# Step 4: Add result column to compare scores
combined['won_chasing'] = combined['team_total_chased'] > combined['team_total_target']

# Optional: Add match date and venue for context
combined = combined.merge(matches[['id', 'date', 'venue']], left_on='match_id', right_on='id')

# Clean final output
# Clean final output
result = combined[['match_id', 'date', 'venue',
                   'batting_team_target', 'team_total_target',
                   'batting_team_chased', 'team_total_chased',
                   'won_chasing']]

print(result)
```
